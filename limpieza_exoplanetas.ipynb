```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
exoplanetas = pd.read_csv('exoplanetas_final.csv')
tierra = pd.read_csv('tierra.csv')
exoestrellas = pd.read_csv('estrellas_exoplanetas.csv')
```

```python
num_cols = ['masa_exoplaneta_m_tierra', 'radio_exoplaneta_r_tierra', 'insolacion', 'temperatura_media_k', 'periodo_orbital_dias', 'excentricidad_orbital']

# Normalizar (0–1) para compararlos bien
scaler = MinMaxScaler()
df_scaled = exoplanetas.copy()
df_scaled[num_cols] = scaler.fit_transform(exoplanetas[num_cols])

# Valores de referencia de la Tierra normalizados también
tierra_ref = scaler.transform([[1, 1, 1, 1, 1, 0]])[0]
# (masa=1, radio=1, insolacion=1, temp=1, periodo=1, excentricidad≈0)

# Calcular cercanía
df_scaled['parecido_tierra'] = ((df_scaled[num_cols] - tierra_ref) ** 2).sum(axis=1)

# Ordenar por los más parecidos (menor distancia)
df_final = df_scaled.sort_values('parecido_tierra').head(20)  # ← cambia 50 si quieres más/menos

```

```python
exoestrellas["nombre_estrella"] = exoestrellas["nombre_estrella"].str.strip().str.upper()
```

```python
exoestrellas_filtrar = [
    "LP 890-9", "TOI-2136", "TOI-2088", "TOI-2093", "K2-285",
    "HD 224018", "TOI-260", "TOI-1260", "HD 110067", "K2-415",
    "TOI-1231", "TOI-1468", "TOI-2322", "TRAPPIST-1", "TOI-178",
    "GJ 3929", "TOI-2345"
]
```

```python
exoestrellas_filtrar = [e.upper() for e in exoestrellas_filtrar]
```

```python
exoestrellas_filtradas = exoestrellas[exoestrellas["nombre_estrella"].isin(exoestrellas_filtrar)]
```

```python
exoestrellas_filtradas = exoestrellas_filtradas.drop_duplicates(subset=["nombre_estrella"])
```
